```python
# Código en Python para "Shaman Tools"


# Base de datos para usuarios y productos
usuarios = {
    "admin": {"password": "123", "rol": "admin"},
    "vendedor1": {"password": "abc", "rol": "vendedor"}
}

productos = {
    "1": {"name": "Ladrillo", "buy_price": 100, "sell_price": 150, "stock": 50, "unit": "unidad"},
    "2": {"name": "Cemento 50kg", "buy_price": 500, "sell_price": 700, "stock": 20, "unit": "saco"}
}

ventas = []

# Función de inicio de sesión
def iniciar_sesion():
    nombre_usuario = input("Ingrese su nombre de usuario: ")
    if nombre_usuario in usuarios:
        contraseña = input("Ingrese su contraseña: ")
        if contraseña == usuarios[nombre_usuario]["password"]:
            print(f"Bienvenido, {nombre_usuario}")
            return nombre_usuario
        else:
            print("Contraseña incorrecta")
    else:
        print("Usuario no encontrado")
    return None

# Función para crear un nuevo usuario (solo admin)
def registrar_usuario():
    nombre_usuario_nuevo = input("Ingrese el nombre de usuario nuevo: ")
    if nombre_usuario_nuevo in usuarios:
        print("El usuario ya existe")
    else:
        contraseña = input("Ingrese la contraseña para el nuevo usuario: ")
        rol = input("Ingrese el rol (admin o vendedor): ")
        usuarios[nombre_usuario_nuevo] = {"password": contraseña, "rol": rol}
        print("Usuario creado exitosamente")

# Función para consultar el inventario
def consultar_inventario():
    codigo_producto = input("Ingrese el código del producto (o deje en blanco para ver todos): ")
    if codigo_producto:
        if codigo_producto in productos:
            print(productos[codigo_producto])
        else:
            print("Producto no encontrado")
    else:
        for producto in productos.values():
            print(producto)

# Función para agregar o actualizar un producto (solo admin)
def agregar_actualizar_producto():
    codigo_producto = input("Ingrese el código del producto: ")
    if codigo_producto in productos:
        cantidad = int(input("Ingrese la cantidad a agregar: "))
        productos[codigo_producto]["stock"] += cantidad
        print("Cantidad de producto actualizada")
    else:
        nombre_producto = input("Ingrese el nombre del producto: ")
        precio_compra = float(input("Ingrese el precio de compra: "))
        precio_venta = float(input("Ingrese el precio de venta: "))
        cantidad = int(input("Ingrese la cantidad: "))
        tipo_unidad = input("Ingrese el tipo de unidad: ")
        productos[codigo_producto] = {
            "name": nombre_producto,
            "buy_price": precio_compra,
            "sell_price": precio_venta,
            "stock": cantidad,
            "unit": tipo_unidad
        }
        print("Producto agregado al inventario")

# Función para vender un producto (solo vendedor)
def vender_producto():
    codigo_producto = input("Ingrese el código del producto: ")
    cantidad = int(input("Ingrese la cantidad a vender: "))
    if codigo_producto in productos:
        if productos[codigo_producto]["stock"] >= cantidad:
            productos[codigo_producto]["stock"] -= cantidad
            total = cantidad * productos[codigo_producto]["sell_price"]
            ventas.append({"codigo_producto": codigo_producto, "cantidad": cantidad, "total": total})
            print(f"Venta realizada con éxito. Total: ${total}")
        else:
            print("Stock insuficiente")
    else:
        print("Producto no encontrado")

# Función para listar ventas (solo admin)
def listar_ventas():
    for venta in ventas:
        print(venta)

# Programa principal
while True:
    print("\nBienvenido a Shaman Tools")
    print("1. Iniciar Sesión\n2. Salir")
    opcion_principal = input(">>> ")

    if opcion_principal == "1":
        usuario_actual = iniciar_sesion()
        if usuario_actual:
            rol_usuario = usuarios[usuario_actual]["rol"]
            if rol_usuario == "admin":
                while True:
                    print("\nMenú Admin")
                    print("1. Registrar Usuario\n2. Agregar/Actualizar Producto\n3. Consultar Inventario\n4. Listar Ventas\n5. Cerrar Sesión")
                    opcion = input(">>> ")

                    if opcion == "1":
                        registrar_usuario()
                    elif opcion == "2":
                        agregar_actualizar_producto()
                    elif opcion == "3":
                        consultar_inventario()
                    elif opcion == "4":
                        listar_ventas()
                    elif opcion == "5":
                        print("Cerrando sesión...")
                        break
                    else:
                        print("Opción no válida")
            elif rol_usuario == "vendedor":
                while True:
                    print("\nMenú Vendedor")
                    print("1. Consultar Inventario\n2. Vender Producto\n3. Cerrar Sesión")
                    opcion = input(">>> ")

                    if opcion == "1":
                        consultar_inventario()
                    elif opcion == "2":
                        vender_producto()
                    elif opcion == "3":
                        print("Cerrando sesión...")
                        break
                    else:
                        print("Opción no válida")
    elif opcion_principal == "2":
        print("Gracias por su visita. ¡Hasta pronto!")
        break
    else:
        print("Opción no válida")
```
